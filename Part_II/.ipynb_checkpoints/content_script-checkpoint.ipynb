# Data loading

```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 200
data = pd.read_csv('content_train.tsv', sep='\t')
data.info()
data.describe(include='all')
```

# Data manipulation

```python
# make data tidy
data = data.set_index('customer.id')

# remove redundant target variables and NA values
redundant_target_vars = ['content_2', 'content_3', 'content_4', 'content_5', 'content_6', 'content_7', 'content_8', 'content_9']
data = data.drop(columns=redundant_target_vars)
data = data.dropna()

# ensure target var type is proper
data['content_1'] = data['content_1'].astype(int)

# transform categoricals into dummies
data = pd.get_dummies(data, drop_first=True)
```

# Data exploration

```python
# illustrate discriminative power of regressors (excluding county)
ordinal_vars = list(data.select_dtypes('float64').columns.values) + list(data.select_dtypes('int64').columns.values) + ['gender_Male', 'affluency_Low', 'affluency_Mid', 'affluency_Very High', 'affluency_Very Low'] 
ordinal_vars.remove('content_1')
data.groupby('content_1').mean()[ordinal_vars]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cols_to_plot = list(data.select_dtypes('float64').columns.values) + list(data.select_dtypes('int64').columns.values)
sns.set()
g = sns.PairGrid(data[cols_to_plot])
g = g.map_diag(plt.hist, bins=20)
g = g.map_offdiag(plt.scatter)
```

# Model training

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import log_loss
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

X = data.drop('content_1', axis=1).values
y = data['content_1'].values

n_splits = 10
k_fold = StratifiedKFold(n_splits=n_splits)

n_trees_array = np.arange(1, 51)
cross_val_score = pd.Series(index=n_trees_array)
cross_val_score_naive = pd.Series(index=n_trees_array)

for n_trees in n_trees_array:    
    test_error = np.zeros(n_splits)
    test_error_naive = np.zeros(n_splits)
    for k, (train, test) in enumerate(k_fold.split(X, y)):
        X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]  
        scaler = StandardScaler().fit(X_train) 
        X_train = scaler.transform(X_train)
        X_test = scaler.transform(X_test)    

        random_forest = RandomForestClassifier(max_depth=2, random_state=0, n_estimators=n_trees)
        random_forest.fit(X_train, y_train)
        
        y_pred = random_forest.predict_proba(X_test)
        y_pred_naive = y_train.mean() * np.ones(len(X_test))

        test_error[k] = log_loss(y_test, y_pred, 1e-15)
        test_error_naive[k] = log_loss(y_test, y_pred_naive, 1e-15)
    
    cross_val_score[n_trees] = test_error.mean()
    cross_val_score_naive[n_trees] = test_error_naive.mean()
    
    print('mean test error random forest: ', test_error.mean())
    print('standard deviation test error random forest: ', test_error.std())
    print('mean test error naive model: ', test_error_naive.mean())
    print('standard deviation test error naive model: ', test_error_naive.std())
    print('')
```

```python
plt.figure()
plt.plot(cross_val_score)
plt.plot(cross_val_score_naive)
plt.xlabel('number of trees')
plt.ylabel('log loss')
plt.legend(('random forest', 'naive'))
```

# Prediction

```python
data_train = pd.read_csv('content_train.tsv', sep='\t')
redundant_target_vars = ['content_2', 'content_3', 'content_4', 'content_5', 'content_6', 'content_7', 'content_8', 'content_9']
data_train = data_train.drop(columns=redundant_target_vars)
data_train = data_train.dropna()
data_test = pd.read_csv('content_test.tsv', sep='\t')
data = pd.concat([data_train, data_test], axis=0)

data = data.set_index('customer.id')
data = pd.get_dummies(data, drop_first=True)

data_train = data.loc[1:100000]
data_test = data.loc[100001:120000]

X_train = data_train.drop('content_1', axis=1).values
y_train = data_train['content_1'].values
X_test = data_test.drop('content_1', axis=1).values

scaler = StandardScaler().fit(X_train) 
X_train = scaler.transform(X_train)

random_forest = RandomForestClassifier(max_depth=2, random_state=0, n_estimators=25)
random_forest.fit(X_train, y_train)

X_test = scaler.transform(X_test) 
y_pred = random_forest.predict_proba(X_test)

df = pd.DataFrame(y_pred[:, 1], index=data_test.index)
df.columns = ['content_1']
df.to_csv('predictions.tsv', sep='\t')
```
